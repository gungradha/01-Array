```python
import numpy as np
```

Array nilai 10 mahasiswa

```python
nilaimhs = np.array([89, 90, 52, 93, 80, 82, 95, 60, 78, 57])
```

Nilai Tertinggi Mahasiswa

```python
nilai_tertinggi = np.max(nilaimhs)
print(nilaimhs)
print("Nilai Tertinggi Mahasiswa: ", nilai_tertinggi)
```

Nilai Terendah Mahasiswa

```python
nilai_terendah = np.min(nilaimhs)
print(nilaimhs)
print("Nilai Terendah Mahasiswa: ", nilai_terendah)
```

Jumlah Mahasiswa Lulus >=60

```python
nilaihms = np.array([89, 90, 52, 93, 80, 82, 95, 60, 78, 57])

jumlah_lulus = np.sum(nilaihms >= 60)
print("Jumlah Mahasiswa Lulus (>=60):", jumlah_lulus)
```
