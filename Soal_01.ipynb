```python
import numpy as np

print("Nilai Mahasiswa :")
nilaimhs = np.array([89, 90, 52, 93, 80, 82, 95, 60, 78, 57])
print(nilaimhs)
```

Nilai Tertinggi & Nilai Terendah Mahasiswa

```python
print("Nilai Mahasiswa :")
print(nilaimhs)

nilai_tertinggi = np.max(nilaimhs)
print("Nilai Tertinggi Mahasiswa: ", nilai_tertinggi)

nilai_terendah = np.min(nilaimhs)
print("Nilai Terendah Mahasiswa: ", nilai_terendah)
```

Nilai Rata-Rata Mahasiswa

```python
rata_rata = np.mean(nilaimhs)
print("Nilai Rata-Rata Mahasiswa :", rata_rata)
```

Jumlah Mahasiswa Lulus >=60

```python
nilaihms = np.array([89, 90, 52, 93, 80, 82, 95, 60, 78, 57])

jumlah_lulus = np.sum(nilaimhs >= 60)
print("Jumlah Mahasiswa Lulus (>=60):", jumlah_lulus)

jumlah_tidak_lulus = np.sum(nilaihms < 60)
print("Jumlah Mahasiswa Tidak Lulus (<60):", jumlah_tidak_lulus)
```

Grafik Nilai Tertinggi dan Nilai Terendah

```python
import matplotlib.pyplot as plt
plt.bar(range(len(nilaimhs)), nilaimhs)
plt.title("Nilai Mahasiswa")
plt.xlabel("Mahasiswa")
plt.ylabel("Nilai")
plt.show()
```

Grafik Data Lulus

```python
import matplotlib.pyplot as plt


labels = ["Lulus", "Tidak Lulus"]
sizes = jumlah_lulus, jumlah_tidak_lulus
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=180)
plt.title("Data Lulus Mahasiswa")
plt.axis('equal')
plt.show()
```
